```python
from google.colab import files
import zipfile, os

# Upload zip manually
uploaded = files.upload()

# Extract
zip_path = "Mask-dataset.zip"   # same name as uploaded
extract_path = "/content/Mask-dataset"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("✅ Dataset extracted at:", extract_path)

# Check structure
for root, dirs, files in os.walk(extract_path):
    print(root, len(files), "files")

import os

# Root after extraction
root_path = "/content/Mask-dataset/Mask-dataset"

# Fix paths
train_img_dir = os.path.join(root_path, "images/train-mask")
val_img_dir   = os.path.join(root_path, "images/val-mask")
train_mask_dir = os.path.join(root_path, "annotations/train_masks")
val_mask_dir   = os.path.join(root_path, "annotations/val_masks")

print("Train images:", train_img_dir)
print("Val images:", val_img_dir)
print("Train masks:", train_mask_dir)
print("Val masks:", val_mask_dir)

# Check counts
print("✅ Train imgs:", len(os.listdir(train_img_dir)))
print("✅ Val imgs:", len(os.listdir(val_img_dir)))
print("✅ Train masks:", len(os.listdir(train_mask_dir)))
print("✅ Val masks:", len(os.listdir(val_mask_dir)))
# =========================================
# 0. Setup - Upload dataset
# =========================================
from google.colab import files
import zipfile, os

# Upload your Mask-dataset.zip
uploaded = files.upload()

zip_path = list(uploaded.keys())[0]  # e.g., Mask-dataset.zip
extract_path = "/content/Mask-dataset"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("✅ Dataset extracted at:", extract_path)

# =========================================
# 1. Check structure
# =========================================
for root, dirs, files in os.walk(extract_path):
    print(root, len(files), "files")

# =========================================
# 2. Dataset Loader
# =========================================
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import torch

class CrackDataset(Dataset):
    def __init__(self, img_dir, mask_dir):
        self.img_dir = img_dir
        self.mask_dir = mask_dir
        self.images = sorted([f for f in os.listdir(img_dir) if f.endswith(('.jpg','.png','.jpeg'))])

        self.img_transform = transforms.Compose([
            transforms.Resize((256, 256)),
            transforms.ToTensor()
        ])
        self.mask_transform = transforms.Compose([
            transforms.Resize((256, 256)),
            transforms.ToTensor()
        ])

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_name = self.images[idx]
        mask_name = os.path.splitext(img_name)[0] + ".png"

        img_path = os.path.join(self.img_dir, img_name)
        mask_path = os.path.join(self.mask_dir, mask_name)

        image = Image.open(img_path).convert("RGB")
        mask  = Image.open(mask_path).convert("L")

        image = self.img_transform(image)
        mask  = self.mask_transform(mask)

        return image, mask

# Fix paths (your dataset)
root_path = "/content/Mask-dataset/Mask-dataset"

train_img_dir = os.path.join(root_path, "images/train-mask")
val_img_dir   = os.path.join(root_path, "images/val-mask")
train_mask_dir = os.path.join(root_path, "annotations/train_masks")
val_mask_dir   = os.path.join(root_path, "annotations/val_masks")

train_dataset = CrackDataset(train_img_dir, train_mask_dir)
val_dataset   = CrackDataset(val_img_dir, val_mask_dir)

train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=2, shuffle=False)

print(f"✅ Train samples: {len(train_dataset)}")
print(f"✅ Val samples: {len(val_dataset)}")

# =========================================
# 3. UNet Model
# =========================================
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()
        def CBR(in_c, out_c):
            return nn.Sequential(
                nn.Conv2d(in_c, out_c, 3, padding=1),
                nn.ReLU(inplace=True),
                nn.Conv2d(out_c, out_c, 3, padding=1),
                nn.ReLU(inplace=True)
            )
        self.enc1 = CBR(3, 64)
        self.enc2 = CBR(64, 128)
        self.enc3 = CBR(128, 256)
        self.enc4 = CBR(256, 512)
        self.pool = nn.MaxPool2d(2)
        self.up1 = nn.ConvTranspose2d(512, 256, 2, stride=2)
        self.dec1 = CBR(512, 256)
        self.up2 = nn.ConvTranspose2d(256, 128, 2, stride=2)
        self.dec2 = CBR(256, 128)
        self.up3 = nn.ConvTranspose2d(128, 64, 2, stride=2)
        self.dec3 = CBR(128, 64)
        self.outc = nn.Conv2d(64, 1, 1)

    def forward(self, x):
        e1 = self.enc1(x)
        e2 = self.enc2(self.pool(e1))
        e3 = self.enc3(self.pool(e2))
        e4 = self.enc4(self.pool(e3))
        d1 = self.up1(e4)
        d1 = torch.cat([d1, e3], dim=1)
        d1 = self.dec1(d1)
        d2 = self.up2(d1)
        d2 = torch.cat([d2, e2], dim=1)
        d2 = self.dec2(d2)
        d3 = self.up3(d2)
        d3 = torch.cat([d3, e1], dim=1)
        d3 = self.dec3(d3)
        return torch.sigmoid(self.outc(d3))

# =========================================
# 4. Training
# =========================================
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

model = UNet().to(device)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

num_epochs = 5
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    for imgs, masks in train_loader:
        imgs, masks = imgs.to(device), masks.to(device)
        preds = model(imgs)
        loss = criterion(preds, masks)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss/len(train_loader):.4f}")

torch.save(model.state_dict(), "/content/unet_model.pth")
print("✅ Training complete, model saved as unet_model.pth")

# =========================================
# 5. Quick Prediction Visualization
# =========================================
import matplotlib.pyplot as plt

model.eval()
imgs, masks = next(iter(val_loader))
with torch.no_grad():
    preds = model(imgs.to(device)).cpu()

for i in range(len(imgs)):
    plt.figure(figsize=(10,3))
    plt.subplot(1,3,1); plt.imshow(imgs[i].permute(1,2,0)); plt.title("Image")
    plt.subplot(1,3,2); plt.imshow(masks[i][0], cmap="gray"); plt.title("Mask")
    plt.subplot(1,3,3); plt.imshow(preds[i][0]>0.5, cmap="gray"); plt.title("Prediction")
    plt.show()

import matplotlib.pyplot as plt
import math

model.eval()
all_imgs, all_masks, all_preds = [], [], []

with torch.no_grad():
    for imgs, masks in val_loader:
        imgs, masks = imgs.to(device), masks.to(device)
        preds = model(imgs).cpu()

        all_imgs.extend(imgs.cpu())
        all_masks.extend(masks.cpu())
        all_preds.extend(preds.cpu())

# Number of validation samples
n = len(all_imgs)
cols = 3   # Image, Mask, Prediction
rows = n   # one row per sample

fig, axs = plt.subplots(rows, cols, figsize=(12, 4*rows))

for i in range(n):
    axs[i,0].imshow(all_imgs[i].permute(1,2,0))
    axs[i,0].set_title("Image")
    axs[i,0].axis("off")

    axs[i,1].imshow(all_masks[i].squeeze(), cmap="gray")
    axs[i,1].set_title("Mask")
    axs[i,1].axis("off")

    axs[i,2].imshow(all_preds[i].squeeze(), cmap="gray")
    axs[i,2].set_title("Prediction")
    axs[i,2].axis("off")

plt.tight_layout()
plt.show()
import numpy as np

def iou_score(pred, target, threshold=0.5):
    pred = (pred > threshold).float()
    intersection = (pred * target).sum()
    union = pred.sum() + target.sum() - intersection
    if union == 0:
        return torch.tensor(1.0) if intersection == 0 else torch.tensor(0.0)
    return intersection / union

def dice_score(pred, target, threshold=0.5):
    pred = (pred > threshold).float()
    intersection = (pred * target).sum()
    return (2.0 * intersection) / (pred.sum() + target.sum() + 1e-8)

def pixel_accuracy(pred, target, threshold=0.5):
    pred = (pred > threshold).float()
    correct = (pred == target).float().sum()
    total = torch.numel(pred)
    return correct / total

# ======================================
# Evaluate model on validation set
# ======================================
model.eval()
iou_list, dice_list, acc_list = [], [], []

with torch.no_grad():
    for imgs, masks in val_loader:
        imgs, masks = imgs.to(device), masks.to(device)
        preds = model(imgs)

        for i in range(len(imgs)):
            iou_list.append(iou_score(preds[i], masks[i]).item())
            dice_list.append(dice_score(preds[i], masks[i]).item())
            acc_list.append(pixel_accuracy(preds[i], masks[i]).item())

print(f"✅ Mean IoU: {np.mean(iou_list):.4f}")
print(f"✅ Mean Dice: {np.mean(dice_list):.4f}")
print(f"✅ Pixel Accuracy: {np.mean(acc_list):.4f}")
from torchvision import transforms

train_transform = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(20),
    transforms.ToTensor()
])
from torchvision import transforms

train_transform = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(20),
    transforms.ToTensor()
])
```
